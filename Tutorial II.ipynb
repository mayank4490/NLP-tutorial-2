# Introduction to NLTK

```python
import nltk
from nltk.corpus import brown
```

## Brown Corpus 
### Statistics:
    ~1.1 Million tokens
    500 text documents
    15 categories
More details about brown corpus: https://tinyurl.com/browncorpus

Other corpora details can be found at:  http://www.nltk.org/book/ch02.html

```python
tokenList  =  brown.words()
print tokenList
```

```python
print "Total token count:", len(tokenList)
```

```python
print brown.categories()
```

```python
adventureTokenList =  brown.words(categories='adventure')
print adventureTokenList
```

```python
print "Token count of adventure category:", len(adventureTokenList)
```

```python
print "Brown vocabulary size:", len(set(tokenList))  # Also token types
```

```python
print "TTR for Brown corpus:", len(set(tokenList)/ float(len(tokenList))
```

```python
tokenCount = nltk.FreqDist(tokenList)
print tokenCount
```

```python
print tokenCount.most_common(50)
```

```python
tokenCount.plot(50, cumulative=True)
```

```python
from collections import defaultdict
CountFreq = defaultdict(int)
for word in tokenCount:
    CountFreq[tokenCount[word]]+=1
```

```python
print  CountFreq
```

```python
import matplotlib.pyplot as plt

plt.scatter(CountFreq.keys(), CountFreq.values())
plt.xlabel('Token Count')
plt.ylabel('Token Count Frequency')
plt.yscale('log')
plt.xscale('log')
plt.ylim(0,25560)
plt.xlim(0,70000)
plt.show()
```
